```python
def create_file_speeches(url, counter, save_file):
    source_code = requests.get(url, timeout = 120)

    plain_text = source_code.text
    soup = BeautifulSoup(plain_text)
    
    if soup.find('div', class_="transcript-inner"):
        speech = soup.find('div', class_="transcript-inner").find_all('p')
    elif soup.find('div',class_="view-transcript"):
        speech = soup.find('div',class_="view-transcript").find_all('p')
    
    text = ''
    for i in speech:
        text += i.text.strip()
    speaker = soup.find('div', class_="presidential-speeches--body-wrapper").find('label', class_="presidential-speeches--label").text.strip()
    
    date_title = soup.find('h2', class_="presidential-speeches--title").find('span').text.strip()
    split_title = date_title.split(':')
    title = split_title[1]
    date = split_title[0]
    
    date_time = datetime.strptime(date, "%B %d, %Y")
    date_time = date_time.strftime("%m%d%Y")

    File_object = codecs.open(save_file+str(counter)+speaker+date_time+".txt","a", "utf-8")
    File_object.write(text)
    File_object.close()
    
    
```
